### Error analysis
Print out the errors for each context dimension

```python
import os
import json
context_ann_dir = 'data/split_data/output/context_predicted_annotations'
gold_std_ann_dir = 'data/split_data/test'

for model_name in ['action', 'actor', 'temporality', 'certainty', 'negation']:
    print('\n', model_name, '\n**************************************************************************************************************************')

    root_names = [file.replace(".ann", "") for file in 
                  os.listdir(context_ann_dir) if ".ann" in file]

    # input data
    input_path = f'data/split_data/input/context_input/{model_name}_test.json'
    json_file = open(input_path,'r')
    json_lines = [line for line in json_file]

    error_counter = 0
    for root_name in root_names:
      
        new_lines = []
        ann_file = os.path.join(context_ann_dir, root_name + ".ann")
        gold_file = os.path.join(gold_std_ann_dir, root_name + ".ann")

        ann = open(ann_file, "r")
        gold_ann = open(gold_file, "r")
        lines = ann.readlines()
        gold_lines = gold_ann.readlines()

        for i, line in enumerate(lines):
            if line.strip().startswith('A'):
                attr_id, attr_m = line.strip().split('\t')
                attr_type, attr_arg, attr_val = attr_m.split(' ')
                # only update the attributes for this specfic model
                if attr_type == model_name.capitalize():
                      _, gold_attr_m = gold_lines[i].strip().split('\t')
                      _, _, gold_attr_val = gold_attr_m.split(' ')
                      if gold_attr_val != attr_val:
                          error_counter += 1
                          if error_counter > 10: 
                              break
                          print('***************************************************************************')
                          print("\nError ", error_counter)
                          print("Predicted value: ", attr_val)
                          print("Actual value: ", gold_attr_val)

                          # find matching text from json
                          for json_line in json_lines:

                              input_json = json.loads(json_line)
                              if input_json['note_id'] == root_name:
                                  if input_json['rid'] == attr_id:
                                      print("Text: \n", input_json['text'])
                                      break
```

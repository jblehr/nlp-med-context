```python
from sklearn.model_selection import train_test_split
import os
import shutil
```

If you decide to re-run the function below with a new test_size or new random_state, be sure to delete the old directories and create new ones.

"data/split_data/train"
"data/split_data/dev"
"data/split_data/test"

```python
def split_train_dev_test(test_size=0.1, random_state=5):
    '''
    This function splits the existing .ann and .txt files from
    "data/training_v3/train" and "data/training_v3/dev" into
    train, dev, and test sets. The files are copied over from
    the old directories into the following directories:
    "data/split_data/train"
    "data/split_data/dev"
    "data/split_data/test"

    Inputs:
        test_size (float) - the percent of files to hold out
                            for the test set
        random_state (int) - the random state to be used in
                            sklearn's train_test_split()
    Returns:
        Nothing. Just copies the appropriate files into new
        directories.
    '''
    base_dir = 'data/trainingdata_v3'

    for data_split in ["train", "dev"]:
        print(f"Splitting {data_split}")
        data_dir = os.path.join(base_dir, data_split)
        root_names = [file.replace(".ann", "") for file in 
                    os.listdir(data_dir) if ".ann" in file]
        print("Total original files:", len(root_names))

        # Split dataset into training set and test set
        root_names_train, root_names_test = train_test_split(root_names, 
                                                             test_size=test_size,
                                                             random_state=random_state)
        print(f"Total {data_split} files:", len(root_names_train))
        print("Total test files:", len(root_names_test))

        for root_name in root_names_train:
            for f_type in [".ann", ".txt"]:
                src = os.path.join(data_dir, root_name) + f_type
                dst = "data/split_data/" + data_split
                dst = os.path.join(dst, root_name) + f_type
                shutil.copyfile(src, dst)
        
        for root_name in root_names_test:
            for f_type in [".ann", ".txt"]:
                src = os.path.join(data_dir, root_name) + f_type
                dst = "data/split_data/test"
                dst = os.path.join(dst, root_name) + f_type
                shutil.copyfile(src, dst)
```

```python
split_train_dev_test()
```
